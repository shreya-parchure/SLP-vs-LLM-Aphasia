```python
!pip install pandas openpyxl scipy matplotlib seaborn statsmodels
import pandas as pd
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.stats.multicomp import pairwise_tukeyhsd

file_path = "allData.xlsx"
df = pd.read_excel(file_path, sheet_name="linearRegression")

# rename columns
df = df.rename(columns={
    "WAB 1 AQ": "WAB AQ",
    "WAB Naming subscore": "WAB Naming Subscore",
    "# AI Hallucinations": "AI Hallucinations",
    "% Match": "Match Rate",
    "Aphasia Type": "Aphasia Type"
})

# define predictor and outcome
# here we examine whether WAB Naming Subscore predicts match rate. adjust variables to examine other linear regression relationships
X = df['WAB Naming Subscore'] # predictor
y = df['Match Rate'] # dependent variable
X = sm.add_constant(X)

# fit linear regression model
model = sm.OLS(y, X).fit()
print("=== Linear Regression Summary ===")
print(model.summary())

# regression plot, shows relationship between naming subscore and match rate,
# including a best-fit regression line with a 95% confidence interval.
sns.regplot(x='WAB Naming Subscore', y='Match Rate', data=df, ci=95, scatter_kws={'alpha':0.7})
plt.title('Relationship Between WAB Naming Subscore and Match Rate')
plt.xlabel('WAB Naming Subscore')
plt.ylabel('Match Rate (%)')
plt.tight_layout()
plt.show()
```
