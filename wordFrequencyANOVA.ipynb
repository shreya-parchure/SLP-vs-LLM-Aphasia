```python
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# load excel file
file_path = "allData.xlsx"
df = pd.read_excel(file_path, sheet_name="wordFrequency")

# define condition groups based on word frequency
high = df.loc[df['frequency'] == 'H', 'match rate']
medium = df.loc[df['frequency'] == 'M', 'match rate']
low = df.loc[df['frequency'] == 'L', 'match rate']

f_stat, p_value = stats.f_oneway(high, medium, low)
print("=== One-way ANOVA ===")
print(f"F-statistic = {f_stat:.4f}")
print(f"p-value = {p_value:.4e}\n")

# post-hoc analysis
# run Tukey's HSD test
tukey = pairwise_tukeyhsd(
    endog=df['error rate (match rate?)'],
    groups=df['frequency'],
    alpha=0.05
)
print("=== Tukey HSD Post-hoc Test ===")
print(tukey, "\n")
```
